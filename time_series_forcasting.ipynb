```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
date_rng = pd.date_range(start= "2023-01-02", end ="2023-12-31", freq ="D" )
```

```python
date_rng
```

```python
time_series_data = pd.Series(np.random.randn(len(date_rng)), index = date_rng)
```

```python
time_series_data
```

```python
plt.plot(time_series_data)
plt.title("ARIMA data")
plt.show()
```

```python
model = ARIMA(time_series_data, order = (1,1,1))
```

```python
(2,1,3)
```

```python
fit_model = model.fit()
```

```python
forecast_steps = 30
forecast = fit_model.forecast(steps = forecast_steps)
```

```python
plt.plot(time_series_data, label = "Original Data")
plt.plot(forecast, label = "Forecast")
plt.legend()
plt.title("ARIMA forecasts")
plt.show()
```
